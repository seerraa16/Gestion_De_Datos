```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos (ajusta la ruta según tu archivo)
df = pd.read_csv('tabla.csv')

# Estilo de los gráficos
sns.set_style("whitegrid")

# 1. Histograma de la edad de los clientes
plt.figure(figsize=(8,5))
sns.histplot(df['Edad'], bins=20, kde=True, color='royalblue')
plt.title('Distribución de Edad de los Clientes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

# 2. Gráfico de barras del churn estimado
plt.figure(figsize=(8,5))
sns.histplot(df['churn_estimado'], bins=20, kde=True, color='red')
plt.title('Distribución del Churn Estimado')
plt.xlabel('Churn Estimado')
plt.ylabel('Frecuencia')
plt.show()

# 3. Gráfico de barras de retención estimada
plt.figure(figsize=(8,5))
sns.histplot(df['retencion_estimado'], bins=20, kde=True, color='green')
plt.title('Distribución de la Retención Estimada')
plt.xlabel('Retención Estimada')
plt.ylabel('Frecuencia')
plt.show()

# 4. Boxplot de CLTV a lo largo de los años
cltv_columns = ['CLTV_1_anio', 'CLTV_2_anios', 'CLTV_3_anios', 'CLTV_4_anios', 'CLTV_5_anios']
df_melted = df.melt(value_vars=cltv_columns, var_name='Periodo', value_name='CLTV')

plt.figure(figsize=(10,6))
sns.boxplot(x='Periodo', y='CLTV', data=df_melted, palette='coolwarm')
plt.title('Distribución del CLTV por Año')
plt.xlabel('Período de CLTV')
plt.ylabel('CLTV')
plt.show()

# 5. Relación entre edad y CLTV_5_anios
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Edad'], y=df['CLTV_5_anios'], alpha=0.5, color='purple')
plt.title('Relación entre Edad y CLTV a 5 años')
plt.xlabel('Edad')
plt.ylabel('CLTV a 5 años')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df = pd.read_csv("tabla.csv")

# Asegurarse de que los valores estén en formato numérico
df['CLTV_5_anios'] = df['CLTV_5_anios'].replace(',', '.', regex=True).astype(float)

# Calcular la media
media = df['CLTV_5_anios'].mean()

# Crear el gráfico
plt.figure(figsize=(10, 6))
sns.histplot(df['CLTV_5_anios'], kde=True, bins=60, color='cornflowerblue', edgecolor='black', alpha=0.6)
plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: €{media:,.0f}')
plt.title('Distribución del CLTV a 5 Años')
plt.xlabel('CLTV Acumulado (€)')
plt.ylabel('Frecuencia')
plt.legend()
plt.tight_layout()
plt.show()
```
